## **PROBLEM STATEMENT**

Exploratory data analysis is required to be conducted on the population density (i.e. the number of people per square kilometer). 
The focus of this exploration is on countries with the highest population densities. 

# **Import libraries** 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# **Importing the data** 

```python
data = pd.read_csv('https://tinyurl.com/ChrisCoDV/world_population.csv',index_col=1)
pd.plotting.register_matplotlib_converters()
data
```

```python
data =data.transpose()
selected = ['Country Name', 'Indicator Name','Indicator Code', '1960','2016'] #The columns to be dropped
data = data.drop(selected, axis=0)
```

```python
data.index = pd.to_datetime(data.index) #set date time
```

```python
print(data.head()) #To confirm that the Country Name, Indicator Name,Indicator Code and 1960 are all dropped
print(data.tail()) #To confirm 2016 has been dropped
```

# **Line plot for all countries** 


```python
data = data.reindex(data.sum().sort_values(ascending=False).index, axis=1) #ordering the data columns by volume(largest first)
data
```

```python
plt.figure(figsize=(8, 8))
plt.plot(data, linewidth=0.5)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Population Density', fontsize=18)
plt.title('Population Density of all countries', fontsize=20)
plt.legend(data.columns, loc=2)
plt.show()

#The legend shows the reordered data, starting with countries with the highets population densities which are MAC,MCO,HKG,SGP and	GIB
```

# **Line plot for the top five countries by population density** 

```python
selected = ['MAC', 'MCO', 'HKG','SGP', 'GIB']

plt.figure(figsize=(8, 8))
plt.plot(data[selected], linewidth=1)
plt.ylim(ymin=0)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Population Density', fontsize=18)
plt.title('Population Density of the top five countries', fontsize=20)
plt.legend(selected, loc=2)
plt.show()
```

```python
period = 10
rolling_average = data.rolling(window=period).mean()

plt.figure(figsize=(8, 8))
plt.plot(data[selected], linewidth=0.5)
plt.gca().set_prop_cycle(None)
plt.plot(rolling_average[selected], linewidth=2)
plt.ylim(ymin=0)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Population Density', fontsize=18)
plt.title('Population Density of the top five countries', fontsize=20)
plt.legend(selected, loc=2)
plt.show()

#The data is relatively clear, hence rolling averages does little to improve understanding the data.
```

# **Line plot with trendlines for the top five countries by population density** 

```python
plt.figure(figsize=(8, 8))
plt.plot(data[selected], linewidth=0.5)
plt.gca().set_prop_cycle(None)
for name in selected:
    x = np.arange(len(data[name]))
    z = np.polyfit(x, data[name].astype(np.float64), 1)
    trend = np.poly1d(z)
    plt.plot(data.index, trend(x), linestyle='--')
plt.ylim(ymin=0)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Population Density', fontsize=18)
plt.title('Population Density of the top five countries', fontsize=20)
plt.legend(selected, loc=2)
plt.show()

#The data is relatively clear, hence Trendlines does little to improve understanding the data.
```
